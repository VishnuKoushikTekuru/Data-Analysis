# SQL query to find the number of doctors that contributed to the state having highest 'cash' fine

```python
import csv, sqlite3

con = sqlite3.connect("fine_data.db") 
cur = con.cursor()
cur.execute("CREATE TABLE fine_data (org_indiv VARCHAR(255) NOT NULL,first_name VARCHAR(255),last_name VARCHAR(255),city VARCHAR(255),state VARCHAR(255),category VARCHAR(255),cash_fine INTEGER,online_fine INTEGER,name VARCHAR(255));") 

with open('fine_data.csv','r') as csvfile:
    dr = csv.DictReader(csvfile) 
    to_db = [( i['org_indiv'], i['first_name'], i['last_name'], i['city'], i['state'], i['category'], i[' cash_fine '], i[' online_fine '], i['name']) for i in dr]

cur.executemany("INSERT INTO fine_data( org_indiv, first_name,last_name ,city ,state , category ,cash_fine ,online_fine ,name) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?);", to_db)
cur.execute("SELECT * FROM fine_data")
print(cur.fetchall())
```

```python
print(cur.execute("select count(name) from fine_data ").fetchall())
```

```python
print(cur.execute("select state,name, sum(cash_fine) as total_cash from fine_data group by state ").fetchall())
```

```python
print(cur.execute("select state from (select state, sum(cash_fine) as total_cash from fine_data group by state) order by total_cash desc").fetchall())
```

```python
print(cur.execute("select count(name),state from fine_data where state = (select state from (select state, sum( cash_fine ) as total_cash from fine_data group by state) order by total_cash desc) and cash_fine").fetchall())
```

shows that there are around 524 Doctors who contributed the to state with highest cash fine
